## ðŸ“Š Exercise 4 â€” Summary stats & visuals (Y vs X)
From `df`, select **one outcome (Y)** and **one predictor (X)**, then:
1. Print `.describe()` for these variables.
2. Plot a histogram for **Y** and a histogram for **X**.
3. Plot a scatter of **Y vs X**.

```python
# your code here
import matplotlib.pyplot as plt

# Example choice:
#   Y = Dec (CPI in December each year)
#   X = Year

Y = df["Dec"]
X = df["Year"]

# 1. Summary stats
print("Summary stats for Y (Dec CPI):")
print(Y.describe())
print("\nSummary stats for X (Year):")
print(X.describe())

# 2. Histograms
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.hist(Y, bins=10, color="skyblue", edgecolor="black")
plt.title("Histogram of December CPI (Y)")
plt.xlabel("CPI Value")
plt.ylabel("Frequency")

plt.subplot(1,2,2)
plt.hist(X, bins=len(X), color="lightgreen", edgecolor="black")
plt.title("Histogram of Years (X)")
plt.xlabel("Year")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()

# 3. Scatter plot
plt.scatter(X, Y, color="purple")
plt.title("December CPI vs Year")
plt.xlabel("Year")
plt.ylabel("December CPI")
plt.show()
```
